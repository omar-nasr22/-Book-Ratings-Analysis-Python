```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Books_Data_Clean.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[df["Publishing Year"] >= 1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset="Book Name", inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df["Publishing Year"])
plt.title("Distribution of Publishing Years")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

```python
df["genre"].value_counts().plot(kind='bar')
plt.title("Distribution of Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x='genre', y='Book_ratings_count', data=df)
plt.title("Book Rating Count by Genre")
plt.xlabel("Genre")
plt.ylabel("Rating Count")
plt.show()
```

```python
plt.scatter(df["sale price"],df["units sold"])
plt.title("Sales Price vs Units Sold")
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.show()
```

```python
language_counts= df['language_code'].value_counts()
```

```python
plt.pie(language_counts, labels=language_counts.index, startangle=90, autopct='%1.1f%%')
plt.title("Distribution of Languages")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
df.groupby('Publisher ')["publisher revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby('Author_Rating')["Book_ratings_count"].mean().sort_values(ascending=False).max()
```

```python
df.groupby('language_code').size().sort_values(ascending=False)
```

```python
df.groupby('Author')["Book_ratings_count"].max()
```

```python
plt.scatter(df["Book_average_rating"], df["Book_ratings_count"])
plt.title("Book Average Rating vs Book Ratings Count")
plt.xlabel("Book Average Rating")
plt.ylabel("Book Ratings Count")
plt.show()
```

```python
total_gross_sales_by_author = df.groupby('Author')['gross sales'].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Authors by Gross Sales")
plt.xlabel("Author")
plt.ylabel("Gross Sales")
plt.show()
```

```python
sns.boxenplot(x='Author_Rating',y='units sold', data=df)
plt.title("Units Sold by Author Rating")
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.show()
```

```python
df.groupby('Publishing Year')['units sold'].sum().plot(kind='line')
plt.title("Total Units Sold by Year")
plt.xlabel("Year")
plt.ylabel("Total Units Sold")
plt.show()
```
